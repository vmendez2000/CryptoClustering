```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use StandardScaler() to normalize the data from the CSV file
scaler = StandardScaler()
df_market_data_scaled = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(df_market_data_scaled, columns=df_market_data.columns)
```

```python
# Copy the crypto names from the original DataFrame
crypto_names = df_market_data.index
```

```python
# Set the coin_id column as index
df_scaled['coin_id'] = crypto_names
```

```python
# Display the scaled DataFrame
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12)
inertia_values = []
```

```python
# Ensure df_scaled is properly structured (if 'coin_id' is the index, no need to drop it)
features = df_scaled.drop(columns=['coin_id'], errors='ignore')  # Avoid error if 'coin_id' isn't a column
```

```python
# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Explicitly set n_init for compatibility
    kmeans.fit(features)
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k_values, 'inertia': inertia_values}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(x='k', y='inertia', width=800, height=400)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4, this is where the inertia starts to level off. 

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(features)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans.predict(features)
# Print the resulting array of cluster values
print(clusters)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy['cluster'] = clusters
# Display the copy of the scaled DataFrame
df_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot for visualizing clusters
df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    by="cluster", 
    hover_cols=["coin_id"], 
    title="Cryptocurrency Clusters", 
    width=800, height=400
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
df_pca = pca.fit_transform(df_scaled.drop(columns=['coin_id'], errors='ignore'))
# Ensure coin_id is correctly assigned as index without NaNs
df_pca_df = pd.DataFrame(df_pca, columns=['PC1', 'PC2', 'PC3'])
df_pca_df.index = df_scaled['coin_id']
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
```

```python
# Total explained variance
total_explained_variance = explained_variance.sum()
print(f"Total Explained Variance: {total_explained_variance:.4f}")
```

```python
# Display the first few rows of the PCA DataFrame
df_pca_df.head()
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 0.8950.

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12) 
```

```python
# Create an empty list to store the inertia values
inertia_values_pca = []
# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_pca_df)
    inertia_values_pca.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {'k': k_values, 'inertia': inertia_values_pca}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df_pca = pd.DataFrame(elbow_data_pca)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_df_pca.hvplot.line(x='k', y='inertia', width=800, height=400)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k using PCA data is 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** This does not differ from the original data. 

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans_pca = KMeans(n_clusters=3, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(df_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
clusters_pca = kmeans_pca.predict(df_pca_df)
# Print the resulting array of cluster values.
print(clusters_pca)
```

```python
# Create a copy of the PCA DataFrame and add the predicted clusters
df_pca_df_copy = df_pca_df.copy()
df_pca_df_copy['cluster'] = clusters_pca
# Display the copy of the scaled PCA DataFrame
df_pca_df_copy.head()
```

```python
# Create a scatter plot using hvPlot for visualizing PCA clusters
df_pca_df_copy.hvplot.scatter(
    x="PC1", y="PC2", by="cluster", hover_cols=["coin_id"], width=800, height=400)
```
